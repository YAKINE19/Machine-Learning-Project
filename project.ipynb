```python
# ------------------------------- Optional:  Dictionaries Prints --------------------------------------------------


def print_likelihoods(likelihoods):
    """
    Prints the contents of the "likelihoods" dictionary in a user-friendly format.

    Args:
        likelihoods (dict): The "likelihoods" dictionary to be printed.
    """

    for feature, likelihood_dict in likelihoods.items():
        print(f"\nFeature: {feature}")
        for key, value in likelihood_dict.items():
            print(f"  - {key}: {value:.2f}")

# Example usage:
#likelihoods = {
    #"feature1": {"class1": 0.75, "class2": 0.25},
   # "feature2": {"class1": 0.60, "class2": 0.40}
#}

#print_likelihoods(likelihoods)


# _______________________________________________

def print_pred_priors(pred_priors):
    """
    Prints the contents of the "pred_priors" dictionary in a user-friendly format.

    Args:
        pred_priors (dict): The "pred_priors" dictionary to be printed.
    """

    for feature, priors_dict in pred_priors.items():
        print(f"\nFeature: {feature}")
        for value, probability in priors_dict.items():
            print(f"  - {value}: {probability:.2f}")




# _______________________________________________


def print_class_priors(class_priors):
    """
    Prints the contents of the "class_priors" dictionary in a user-friendly format.

    Args:
        class_priors (dict): The "class_priors" dictionary to be printed.
    """

    print("\nClass priors:")
    for class_label, probability in class_priors.items():
        print(f"  - {class_label}: {probability:.2f}")





# _______________________________________________


```

```python
class NaiveBayes:

    def __init__(self):
        self.features = list  # List of feature names
        self.likelihoods = {}  # Likelihood of each feature value given a class
        self.class_priors = {}  # Prior probabilities of each class
        self.pred_priors = {}  # Prior probabilities of each feature value
        self.X_train = np.array  # Training features
        self.y_train = np.array  # Training labels
        self.train_size = int  # Number of training samples
        self.num_feats = int  # Number of features

    def fit(self, X, y):
        self.features = list(X.columns)
        self.X_train = X.values
        self.y_train = y.values
        self.train_size = len(y)
        self.num_feats = X.shape[1]

        for feature in self.features:
            self.likelihoods[feature] = {}
            self.pred_priors[feature] = {}
            for feat_val in np.unique(X[feature]):
                self.pred_priors[feature][feat_val] = 0
                for outcome in np.unique(y):
                    self.likelihoods[feature][f"{feat_val}_{outcome}"] = 0
                    self.class_priors[outcome] = 0

        self._calc_class_prior()
        self._calc_likelihoods()
        self._calc_predictor_prior()

        print("\n -------- Likelihoods Prints: ----------")
        print_likelihoods(self.likelihoods)
        print("\n -------- Pred_Priors Prints: ----------")
        print_pred_priors(self.pred_priors)
        print("\n -------- Class_Priors Prints: ---------")
        print_class_priors(self.class_priors)

    def _calc_class_prior(self):
        for outcome in np.unique(self.y_train):
            outcome_count = np.sum(self.y_train == outcome)
            self.class_priors[outcome] = outcome_count / self.train_size

    def _calc_likelihoods(self):
        for feature in self.features:
            for outcome in np.unique(self.y_train):
                outcome_rows = self.X_train[self.y_train == outcome]
                for feat_val in np.unique(self.X_train[:, self.features.index(feature)]):
                    feat_val_count = np.sum(outcome_rows[:, self.features.index(feature)] == feat_val)
                    self.likelihoods[feature][f"{feat_val}_{outcome}"] = feat_val_count / np.sum(
                        self.y_train == outcome)

    def _calc_predictor_prior(self):
        for feature in self.features:
            for feat_val in np.unique(self.X_train[:, self.features.index(feature)]):
                self.pred_priors[feature][feat_val] = np.sum(
                    self.X_train[:, self.features.index(feature)] == feat_val) / self.train_size

    def predict(self, X):
        results = []
        X = X.values
        for query in X:
            probs_outcome = {}
            for outcome in np.unique(self.y_train):
                prior = self.class_priors[outcome]
                likelihood = 1
                for feat, feat_val in zip(self.features, query):
                    likelihood *= self.likelihoods[feat][f"{feat_val}_{outcome}"]
                posterior = prior * likelihood
                probs_outcome[outcome] = posterior
            print(probs_outcome)
            result = max(probs_outcome, key=probs_outcome.get)
            results.append(result)
        return np.array(results)
```
